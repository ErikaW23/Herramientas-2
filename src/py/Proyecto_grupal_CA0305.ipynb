# Ejercicio 1: Fundamentos del broadcasting en tensores
Para este primer ejercicio, se debe sumar el vector b a cada fila de la matriz X mediante broadcasting y luego multiplicar cada fila resultante por el vector w. En cuanto a principios matemáticos, esto no se trata de una multiplicación de matrices como en la teoría, si no que se utiliza la dimensión de los vectores b y w con el objetivo de multiplicar matrices elemento por elemento. 

Es importante mencionar otro principio matemático, y es que las dimensiones de las matrices y vectores deben coincidir. Por ejemplo, si X tiene dimensiones (N, D), entonces para sumar por cada elemento debemos hacer que el vector que se suma entrada por entrada sea de tamaño (D,). Además, se está utilizando el principio de broadcasting, que es un mecanismo de NumPy para aperar entre arreglos sin copiar los datos manualmente. 

Para la solución, se reciben como parámetros una matriz X de tamaño (N, D) y dos vectores: b de tamaño (D, ) y w de tamaño (N, ). Dado que los datos pueden ingresarse inicialmente como listas de Python, primero se convierten a arreglos de NumPy mediante np.array(), para las operaciones vectorizadas.

Una vez convertidos, el vector b tiene el tamaño adecuado para sumarse a cada columna de la matriz X. Utilizando el mecanismo de broadcasting de NumPy, esta suma se realiza directamente, ya que NumPy expande automáticamente el vector b a todas las filas de la matriz sin necesidad de replicarlo manualmente.

Posteriormente, la forma original del vector w de tamnaño (N, ) no es compatible con una matriz de tamaño (N, D) para multiplicación elemento a elemento por filas. Por tanto, es necesario convertirlo a una forma (N, 1) utilizando reshape(-1,1). Finalmente, se realiza la operación (X + b) * w, obteniendo una nueva matriz de tamaño (N, D).

```python
import numpy as np

def broadcast_ops(X, b, w):
    X = np.array(X)
    b = np.array(b)
    w = np.array(w)
    
    mult = (X + b) * w.reshape(-1,1)
    return mult

#Ejemplo y prueba sugerida provenientes de la página de PaperCode
X1 = [[1, 2], 
     [3, 4]]
print(broadcast_ops(X1, [10, 20], [0.5, 2]))
```

# Ejercicio 2: Multiplicación de matrices 
Para este ejercicio, se debe calcular el producto entre dos matrices A y B, obteniendo una nueva matriz C. A diferencia del ejercicio anterior, aquí sí se trabaja con la multiplicación matricial que se conoce de la teoría, donde cada elemento de la matriz resultante se obtiene multiplicando una fila de la primera matriz por una columna de la segunda y sumando los productos correspondientes entrada por entrada.

En cuanto a principios matemáticos, se utiliza la definición formal de multiplicación de matrices, igual como aparece en la página de PaperCode, donde el elemento ubicado en la posición (i,j) de la matriz resultado se calcula como la suma de los productos entre los elementos de la fila i de A y la columna j de B. Esto se expresa matemáticamente como:

$$
C_{ij} = \sum_{k=0}^{K-1} A_{ik} B_{kj}
$$

Es importante mencionar que para que esta operación sea posible, las dimensiones internas deben coincidir. Es decir, si A tiene tamaño (M,K), entonces B debe tener tamaño (K,N). Esto garantiza que cada fila de la primera matriz tenga la misma cantidad de elementos que cada columna de la segunda matriz, y se devuelve una matriz de tamaño (M, N).

En este caso, se piden dos funciones, una que utilice estructuras for y sin vectorizar, utilizando tres blucles for anidados, y posteriormente una versión vectorizada utilizando funciones de NumPy. Para la solución, se utilizó el "esqueleto" del código que se encuentra en el ejercicio 2 de el módulo de PaperCode, así, se reciben como parámetros dos matrices A y B, que pueden ingresarse inicialmente como arreglos de NumPy de dos dimensiones. Primero se obtienen sus dimensiones mediante .shape, identificando el número de filas y columnas de cada una, y verificando que las dimensiones internas sean compatibles previo a realizar la operación.

En la versión manual (matmul_naive), se inicializa una matriz de ceros de tamaño (M,N) para almacenar el resultado. Luego se utilizan tres ciclos anidados: el primero recorre las filas de A, el segundo las columnas de B, y el tercero realiza la suma acumulada de los productos correspondientes entre fila y columna para construir cada entrada de la matriz resultado.

Posteriormente, en la versión vectorizada (matmul_vectorized), se utiliza la operación optimizada de NumPy mediante np.dot(), que lleva a cabo el producto matricial.

```python
import numpy as np

#A pesar de que en esta versión no se van a utilizar las funciones de NumPy vectorizadas, si se utiliza la función shape que devuelve el tamaño de la matriz, es por esto que primero en ambos casos es necesario convertir las matrices a obejtos procesables por NumPy como los arreglos. 

def matmul_naive(A: np.ndarray, B: np.ndarray) -> np.ndarray:
    A = np.array(A)
    B = np.array(B)

    M, K = A.shape
    K2, N = B.shape
    assert K == K2 #Verifica que sea verdadero, para ejecutar la multiplicación

    C = np.zeros((M, N))

    for i in range(M):
        for j in range(N):
            for k in range(K):
                C[i,j] += A[i,k] * B[k,j]
    return C

#Funcipon vectorizada
def matmul_vectorized(A: np.ndarray, B: np.ndarray) -> np.ndarray:
    A = np.array(A)
    B = np.array(B)
    return np.dot(A,B)


# Prueba sugerida por la página de PaperCode usando ambas funciones
A = [[1, 2],
     [3, 4]]

B = [[5, 6],
     [7, 8]]

print(matmul_naive(A, B))
print(matmul_vectorized(A, B))
```
